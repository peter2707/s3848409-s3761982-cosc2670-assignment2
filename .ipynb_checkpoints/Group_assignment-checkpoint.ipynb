# Task 1: Retrieving and Preparing the Data

```python
# Importing modules that we will be using
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from statsmodels.stats.weightstats import ztest
###
warnings.filterwarnings('ignore')
###
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
plt.style.use("ggplot")
```

```python
# Load the data from the file heart_failure_clinical_records_dataset.csv using pandas library.
df = pd.read_csv('heart_failure_clinical_records_dataset.csv', sep =',', decimal = '.')
df.head(10)
df1 = df.copy()
```

```python
df['anaemia'] = df['anaemia'].astype(str)
df['diabetes'] = df['diabetes'].astype(str)
df['high_blood_pressure'] = df['high_blood_pressure'].astype(str)
df['sex'] = df['sex'].astype(str)
df['smoking'] = df['smoking'].astype(str)
```

```python
df['anaemia'].replace('1', 'yes', inplace=True)
df['anaemia'].replace('0', 'no', inplace=True)
df['diabetes'].replace('1', 'yes', inplace=True)
df['diabetes'].replace('0', 'no', inplace=True)
df['high_blood_pressure'].replace('1', 'yes', inplace=True)
df['high_blood_pressure'].replace('0', 'no', inplace=True)
df['sex'].replace('1', 'Male', inplace=True)
df['sex'].replace('0', 'Female', inplace=True)
df['smoking'].replace('1', 'yes', inplace=True)
df['smoking'].replace('0', 'no', inplace=True)
```

```python
# checking the feature types and missing values
df.info()
```

```python
# Calculating the statistical significance between averages amongst features with 0.05 significance
for col in df.columns[:-1]:
    # Z test for statistical difference between average DEATH_EVENT by column
    non_death_events = df1[df1['DEATH_EVENT'] == 0][col]
    death_events = df1[df1['DEATH_EVENT'] == 1][col]
    
    ttest, pval = ztest(x1 = non_death_events, x2 = death_events)
    
    if pval <= 0.05:
        print('Hypothesis_Status for %s = Reject Null Hypothesis : Significant'%col)
    else:
        print('Hypothesis_Status for %s = Fail to reject Null Hypothesis : Not Significant'%col)
```

# Task 3: Data Modelling 

```python
# Importing libraries for data modelling
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# Selecting features for data modelling
# Defining feature rows(x) and label row (y)
y = df.DEATH_EVENT
x = df[['ejection_fraction','serum_creatinine','time','serum_sodium','age']]
```

```python
# Set test/train subjects and target variables
# Set test size to 30 %
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

### KNN Classifier

```python
# Generate KNN model and printing out results for evaluation
# And also print confusion matrix
knn  = KNeighborsClassifier(3)
fit = knn.fit(x_train, y_train)
y_pre = fit.predict(x_test)
cm = confusion_matrix(y_test, y_pre)

print("Confusion matrix:\n\n", cm)
print("\n\nClassification Report:\n\n", classification_report(y_test, y_pre))
```

```python
# A for-loop to generate model with n_neighbors starting from 1 to 30
# to find the K value that has the least error rate
error_rate = []

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pre = knn.predict(x_test)
    error_rate.append(np.mean(y_pre != y_test))

min_error_rate = min(error_rate)
min_index = error_rate.index(min(error_rate))
print("Lowest Error Rate:", min_error_rate, "at K Value:", min_index)
```

```python
# A plot to show the relationship betweeen K Value and error rates
# It seems that the least error rate is between 8 to 10
plt.figure(figsize=(10,6))
plt.plot(range(1,30),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='orange', markersize=12)
plt.title('Error Rate vs K Value')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

```python
# We generate the model using the above results to get the best accuracy for KNN Model
knn  = KNeighborsClassifier(8)
fit = knn.fit(x_train, y_train)
y_pre = fit.predict(x_test)
cm = confusion_matrix(y_test, y_pre)

print("Confusion matrix:\n\n", cm)
print("\n\nClassification Report:\n\n", classification_report(y_test, y_pre))
```

### Decision Tree Visualization

```python
# Generating Decision Tree model to evaluate with KNN to find the most accurate model
dt = DecisionTreeClassifier()
fit = dt.fit(x_train, y_train)
y_pre = fit.predict(x_test)
cm = confusion_matrix(y_test, y_pre)

print("Confusion matrix:\n\n", cm)
print("\n\nClassification Report:\n\n", classification_report(y_test, y_pre))
```

```python
#Generating a plot using Graphviz
df['DEATH_EVENT'] = df['DEATH_EVENT'].astype(str)

with open('heart_failure.dot', 'w') as file:
    file = tree.export_graphviz(dt,
                                out_file = file,
                                feature_names = x.columns,
                                class_names = df.DEATH_EVENT,
                                filled=True,
                                rounded=True,
                                special_characters=True)
```
